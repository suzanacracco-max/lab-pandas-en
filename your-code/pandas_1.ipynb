## Let's load another dataset for more exercisesº

```python
# Run this code:
admissions = pd.read_csv('../Admission_Predict.csv')
```

Let's evaluate the dataset by looking at the `head` function.

```python
# your code here
print(admissions.head())
```

### 1 - Before beginning to work with this dataset and evaluating graduate admissions data, we will verify that there is no missing data in the dataset. Do this in the cell below.

```python
# your code here
missing_data = admissions.isnull().sum()
print("Missing values in each column:")
print(missing_data)
```

###  2 -  Interestingly, there is a column that uniquely identifies the applicants. This column is the serial number column. Instead of having our own index, we should make this column our index. Do this in the cell below. Keep the column in the dataframe in addition to making it an index.

```python
# your code here
admissions.set_index("Serial No.", drop=False, inplace=True)

# Display the first 5 rows to verify
display(admissions)
```

Turns out that `GRE Score` and `CGPA` also uniquely identify the data. Show this in the cell below.

```python
# your code here
# Total number of rows in the dataset
total_rows = len(admissions)

# Number of unique combinations of GRE Score and CGPA
unique_combinations = admissions[['GRE Score', 'CGPA']].drop_duplicates().shape[0]

# Check if they uniquely identify each row
is_unique = unique_combinations == total_rows

print("Total rows:", total_rows)
print("Unique GRE + CGPA combinations:", unique_combinations)
print("Do GRE Score and CGPA uniquely identify each row?", is_unique)
```

```python
#other way just testing if GRE Score and CGPA has duplicated values
duplicates_gre = admissions['GRE Score'].duplicated()
print("Duplicates in GRE Score column:")
print(duplicates_gre)

# Check duplicates in CGPA
duplicates_cgpa = admissions['CGPA'].duplicated()
print("\nDuplicates in CGPA column:")
print(duplicates_cgpa)
```

### 3 - In this part of the lab, we would like to test complex conditions on the entire data set at once. Let's start by finding the number of rows where the CGPA is greater than 9 and the student has performed an investigation.

```python
# your code here
# Filter rows where CGPA > 9 and Research == 1
high_cgpa_research = admissions[(admissions['CGPA'] > 9) & (admissions['Research'] == 1)]

# Count the number of rows that satisfy this condition
num_students = high_cgpa_research.shape[0]

print("Number of students with CGPA > 9 and performed research:", num_students)
```

### 4 - Now return all the rows where the CGPA is greater than 9 and the SOP score is less than 3.5. Find the mean chance of admit for these applicants.

```python
# your code here
# Strip any leading/trailing spaces from column names
admissions.columns = admissions.columns.str.strip()

# Filter rows where CGPA > 9 and SOP < 3.5
filtered_applicants = admissions[(admissions['CGPA'] > 9) & (admissions['SOP'] < 3.5)]

# Display the filtered rows
print(filtered_applicants)

# Calculate the mean Chance of Admit for these applicants
mean_admit_chance = filtered_applicants['Chance of Admit'].mean()

print("\nMean Chance of Admit for these applicants:", mean_admit_chance)
```

```python
#or this way
# Filter and calculate mean Chance of Admit in one line using query
mean_admit_chance = admissions.query('CGPA > 9 and SOP < 3.5')['Chance of Admit'].mean()

print("Mean Chance of Admit for applicants with CGPA > 9 and SOP < 3.5:", mean_admit_chance)
```
